# Opción 1

```
from keras.preprocessing import sequence

# maximo número de palabras en el diccionario
max_features = 2000
# recorta el texto a maxlen palabras
maxlen = 10


(x_train_aux, y_train), (x_test_aux, y_test) = keras.datasets.imdb.load_data(num_words=max_features)
print(len(x_train_aux), 'train sequences')
print(len(x_test_aux), 'test sequences')

x_train = sequence.pad_sequences(x_train_aux, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test_aux, maxlen=maxlen)
print(x_train.shape)
print(x_test.shape)
```

# MODELO

```
# Embedding: https://stats.stackexchange.com/questions/270546/how-does-keras-embedding-layer-work

model = Sequential()
model.add(Embedding(max_features,128, input_length=10))
#model.add(Dense(256, activation='relu', input_dim=x_train.shape[1]))
#model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Flatten())
model.add(Dense(1, activation='softmax'))


model.compile(loss='mean_squared_error',
              optimizer='Adam',
              metrics=['binary_accuracy'])
```

```
model.summary()
```

```
y_train.shape
```

```

model.fit(x_train, y_train,
          epochs=200,
          batch_size=128)

```

```

# EVALUACION
score = model.evaluate(x_test, y_test, batch_size=128)

score
```
