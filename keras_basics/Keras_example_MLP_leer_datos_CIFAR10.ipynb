```
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, BatchNormalization
from keras.optimizers import SGD
from keras.datasets import cifar10
    
   
# Coger datos de cifar10

(x_train_aux, y_train_aux), (x_test_aux, y_test_aux) = cifar10.load_data()

x_train = x_train_aux.reshape(x_train_aux.shape[0],x_train_aux.shape[1]*x_train_aux.shape[2]*x_train_aux.shape[3])
x_test = x_test_aux.reshape(x_test_aux.shape[0],x_test_aux.shape[1]*x_test_aux.shape[2]*x_test_aux.shape[3])

from keras.utils import to_categorical
y_train = to_categorical(y_train_aux)
y_test = to_categorical(y_test_aux)
```

```
x_train_aux.shape[1]*x_train_aux.shape[2]*x_train_aux.shape[3]
```

```
# MODELO
model = Sequential()
model.add(Dense(256, activation='relu', input_dim=x_train.shape[1]))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss='mean_squared_error',
              optimizer='Adam',
              metrics=['binary_accuracy'])
```

```
# ENTRENAMIENTO CONTROLANDO EL TIEMPO

import timeit
start = timeit.default_timer()


model.fit(x_train, y_train,
          epochs=40,
          batch_size=128)

stop = timeit.default_timer()
print('Time: ', stop - start)  
```

```

# EVALUACION
score = model.evaluate(x_test, y_test, batch_size=128)

score
```
